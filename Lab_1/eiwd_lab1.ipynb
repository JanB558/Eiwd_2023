Zadanie 40

```python
df1 = pd.read_csv("IHME-GBD_2019_DATA-15798851-2.csv", encoding='utf-8', engine='python')
df2 = pd.read_csv("IHME-GBD_2019_DATA-ff08d9bc-1.csv", encoding='utf-8', engine='python')
```

```python
df1.head()
```

```python
df2.head()
```

```python
df1.rename(columns = {'val': 'val1', 'upper':'upper1', 'lower':'lower1'}, inplace = True)
df2.rename(columns = {'val': 'val2', 'upper': 'upper2', 'lower':'lower2'}, inplace = True)
df_both = pd.merge(df1, df2, on = ['location', 'sex', 'age', 'cause', 'metric', 'year'], how = 'inner') 
df_both.head()
```

```python
#concat
df_part1 = df_both.iloc[:50000,:]
df_part2 = df_both.iloc[50000:,:]
```

```python
df_both_2 = pd.concat([df_part1, df_part2], axis = 0)
df_both_2.head()
```

Zadanie 41

```python
#nowa kolumna będącą zaokrągleniem innej
df_both["val2_round"] = df_both["val2"].round(decimals = 1) 
df_both.head()
```

```python
# nowa kolumna będąca sumą innych kolumn
df_both["sum"] = df_both["val2"] + df_both["upper2"] + df_both["upper1"]
df_both.head()
```

```python
# nowa kolumna poprzez sumowanie wartości kilku innych kolumn metodą loc
df_both["sum2"] = df_both.loc[:, "val2":"lower2"].sum(axis = 1)
df_both.head()
```

```python
# nowa kolumna poprzez sumowanie wartości kilku innych kolumn metodą loc
# od 11 do 13
df_both["sum2_2"] = df_both.iloc[:, 11:14].sum(axis = 1)
df_both.head()
```

```python
df_both.drop(['val2_round','sum', 'sum2','sum2_2'], axis = 1, inplace = True)
```

Zadanie 42

```python
c = ['Poland', 'Czech Republic', 'Italy', 'Sweden', 'Norway', 'Spain', 'Romania']
```

```python
# nowa kolumna c
# jeżeli kraj istnieje w liście c to true, inaczej false
df_both['c'] = df_both['location'].apply(lambda x: True if x in c else False )
df_both[df_both['c'] == True]
```

```python
# krótszy zapis
df_both['c'] = df_both['location'].apply(lambda x: x in c)
```

```python
# wybranie wierszy pasujących do warunku
df_both[df_both['c'] == True]
```

```python
# warunek jako funkcja
def if_c(x):
    if x in c:
        return True
    else:
        False
```

```python
df_both['c'] = df_both['location'].apply(if_c)
```

```python
df_both[df_both['c'] == True]
```

```python
df_both.to_csv('df_both.csv')
# podział dużego pliku przy użycio chunksize

for chunk_df in pd.read_csv('df_both.csv', 
                      chunksize = 50000):
    print("CHUNK DF")
    print(chunk_df.head())
```

```python
# zastosowanie metody groupby oddzielnie do każdej części, a następnie połączenie wyniku w nową ramkę danych
new_df = pd.DataFrame()
for chunk_df in pd.read_csv('df_both.csv', 
                      chunksize = 50000):
    result = chunk_df.groupby(['cause', 'year']).agg({'val2': 'mean',
                                                    'val1': 'mean'})
    new_df = pd.concat([new_df,result])
new_df
```
